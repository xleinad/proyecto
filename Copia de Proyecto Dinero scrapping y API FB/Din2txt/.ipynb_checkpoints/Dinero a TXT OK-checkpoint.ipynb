```python
from bs4 import BeautifulSoup
import requests
import time
import random
import pandas as pd
import numpy as np
```

```python
#Primera nota 2016 66650, última del 2018 94362
lista_ids=list(range(66850,94362)) #lista de ejemplo
```

```python
for ids in lista_ids:  #Este ya sirve, solo que no da salto de linea entre el título y la nota
    start_time =time.time()
    url = "http://www.dineroenimagen.com/node/{}".format(ids)
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')
    titulos = soup.find_all('header', class_='node-title')
    ids = str(ids)+".txt"
    ids = open(ids, "w")
    for tag in titulos:
        if titulos != "":
            titulos = tag.text.strip()+"\n"
            ids.write(titulos)
    notas = soup.find_all('div', class_='field-item even')
    for tag in notas:
        notas = tag.text.strip()
        ids.write(notas)
    ids.close()
    print(round(time.time()-start_time))

```
